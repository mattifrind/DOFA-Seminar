## Running model on images

```python
dataset_path = "data/bigearthnet/"
```

```python
dataset = BigEarthNet(
    root=dataset_path,
    split="train",  # or "val", "test"
    bands="s1",
    num_classes=43,  # or 19 depending on the task
    download=False,
    checksum=False
)
```

```python
dataset.plot(dataset.__getitem__(2000))
```

```python
BATCH_SIZE = 32

dataloader = DataLoader(
    dataset,
    batch_size=BATCH_SIZE,  # Adjust based on your GPU memory
    shuffle=True,
    num_workers=8
)
```

```python
C = 2  # can be 2,3,4,6,9,12,13,202 or any number if you can provide the wavelengths of them
out_feat = torch.zeros(len(dataloader), BATCH_SIZE, 768)
out_logits = torch.zeros(len(dataloader), BATCH_SIZE,  45)
save_labels = torch.zeros(len(dataloader), BATCH_SIZE, 43)
wavelengths = [3.75, 3.75]

with torch.no_grad():
    for index, batch in enumerate(tqdm(dataloader)):
        images = batch["image"]
        labels = batch["label"]
        if images.shape[0] != BATCH_SIZE:
            continue
        preprocessed_images = transform(images)
        preprocessed_images = preprocessed_images.to('cuda')
        out_feat[index] = vit_model.forward_features(preprocessed_images, wave_list=wavelengths).to('cpu')
        out_logits[index] = vit_model.forward(preprocessed_images, wave_list=wavelengths).to('cpu')
        save_labels[index] = labels
        del preprocessed_images
```

```python
torch.save(out_logits, './features/logits-train-bigearthnet.pth')
torch.save(out_feat, './features/feat-train-bigearthnet.pth')
torch.save(save_labels, './features/labels-train-bigearthnet.pth')
```

```python
test_dataset = BigEarthNet(
    root=dataset_path,
    split="test",  # or "val", "test"
    bands="s1",
    num_classes=43,  # or 19 depending on the task
    download=False,
    checksum=False
)
test_dataloader = DataLoader(
    test_dataset,
    batch_size=BATCH_SIZE,  # Adjust based on your GPU memory
    shuffle=True,
    num_workers=8
)
```
